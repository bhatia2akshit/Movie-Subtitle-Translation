```python
import pandas as pd
path = '../data/'
```

```python
import os
dict_languages = {'german':[],'english':[]}
for index in range(1,len(os.listdir(path)), 2):
    files = os.listdir(path)[index]
    print(f'we are at: {files}')
    if files.__contains__('.docx'):
        print(f'processing left for the file {files}')
        continue
    file = open(path+files)
    line = file.readlines()
    german_lines= [line[i] for i in range(0,len(line), 3)]
    english_lines = [line[i] for i in range(1,len(line), 3)]
    
    for german, english in zip(german_lines, english_lines):
        dict_languages['german'].append(german)
        dict_languages['english'].append(english)
```
